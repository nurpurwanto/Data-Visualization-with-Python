## Apa itu Matplotlib?

__Matplotlib__ adalah sebuah <i>library</i> di Python yang sering digunakan untuk visualisasi data. Matplotlib dilengkapi dengan <i>scripting interface</i> yang direpresentasikan dengan __pyplot__ untuk memudahkan membuat grafik. Arsitektur matplotlib terdiri dari tiga layer utama, yaitu <i>backend layer</i>, <i>artist layer</i>, dan <i>scripting layer</i>.

Matplotlib dapat digunakan di banyak <i>tools</i>, salah satunya Google Colab Notebook, yang sedang kita gunakan saat ini.

## Figure dan Axes di Matplotlib

__Figure__ adalah objek Matplotlib yang mengandung semua elemen dari sebuah grafik, sedangkan __Axes__ merupakan tempat dimana <i>method</i> untuk <i>plotting</i> didefinisikan, termasuk <i>method</i> untuk membuat dan memanipulasi garis axis, grid, label tick, ataupun background.

Jika kita asumsikan seperti melukis, Figure adalah kertasnya atau kanvasnya, sedangkan Axes adalah gambarnya. Figure setidaknya mengandung satu Axes. Satu plot (gambar) adalah satu Axes, jadi misalkan membuat sebuah Figure terdiri dari beberapa plot, katakanlah 4 plot, berarti Figure tersebut harus memiliki 4 Axes.

## Membuat plot sederhana

Untuk membuat sebuat plot, impor <code>matplotlib.pyplot</code> terlebih dahulu.

```python
# Import matplotlib.pyplot

import matplotlib.pyplot as plt
```

<code>plt</code> disini adalah sebagai singkatan agar saat dipanggil tidak perlu repot menuliskan <code>matplotlib.pyplot</code> tetapi cukup dengan <code>plt</code>. Sama halnya menyingkat Pandas dengan <code>pd</code> dan Numpy dengan <code>np</code>. 

Ada dua cara yang dapat kita gunakan yaitu cara __Explicit__ dan __Implicit__.

### Cara Explicit

Membuat grafik sederhana dari dua buah kumpulan bilangan yang didefinisikan di dalam dua buah list, yaitu <code>x</code> dan <code>y</code>. 

```python
# Membuat plot sederhana dengan cara explicit

x = [2, 4, 6, 8, 10]
y = [8, 1, 6, 5, 6]

# Membuat objek figure
fig = plt.figure()  

# Membuat objek axes
ax = fig.add_subplot()

# Menambahkan data ke objek axes
ax.plot(x, y)

# Menampilkan visualisasi data
plt.show()
```

Jika menggunakan cara explicit, diharuskan membuat objek figure dan axes terlebih dahulu. Cara ini terbilang sedikit rumit, namun direkomendasikan jika ingin membuat visualisasi data yang lebih kompleks. Dengan cara ini, dapat mengontrol berbagai macam komponen yang terdapat di dalam objek figure dan axes.

Untuk menampilkan visualisasi data dapat memanggil <i>method</i> <code>show()</code> di akhir blok kode, baik dengan cara <i>explicit</i> maupun <i>implicit</i>.

### Cara Implicit

Dengan menggunakan data yang sama, praktekkan penggunaan cara implicit untuk visualisasi data.

```python
# Membuat plot sederhana dengan implicit

x = [2, 4, 6, 8, 10]
y = [8, 1, 6, 5, 6]

# Membuat objek visual 
plt.plot(x, y)

# Menampilkan visualisasi data
plt.show()
```

Dapat kita lihat di atas, kode yang digunakan untuk membuat plot yang sama dengan sebelumnya terlihat lebih sederhana.

## Plotting multiple data dalam satu Axes

Memplot dua data dalam satu Axes. Pada contoh ini, gunakan Numpy untuk men-<i>generate</i> bilangan sehingga perlu mengimpor Numpy terlebih dahulu. Perhatikan kode berikut.

```python
# Plotting multiple data dalam satu Axes

import numpy as np

X = np.linspace(0, 30, 100)
y1 = np.cos(X*2)
y2 = np.cos(X)
y3 = np.sin(X/3)
y4 = np.sin(X)

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(X, y1, color='red')
ax.plot(X, y2, color='blue')
ax.plot(X, y3, color='green') 
ax.plot(X, y4, color='black') 

plt.show()
```

## Mengatur style

Mengatur <i>style</i> di Matplotlib. Untuk mengetahui <i>style</i> apa saja yang tersedia, cek dengan kode di bawah ini.

```python
# Mengecek style yang tersedia

print(plt.style.available)
```

Atur <i>style</i> untuk data trigonometri yang sebelumnya. Misalnya <i>style</i> <code>tableau-colorblind10</code>.

```python
# Mengatur style 

plt.style.use('tableau-colorblind10')

X = np.linspace(0, 30, 100)
y1 = np.cos(X*2)
y2 = np.cos(X)
y3 = np.sin(X/3)
y4 = np.sin(X)

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(X, y1)
ax.plot(X, y2)
ax.plot(X, y3) 
ax.plot(X, y4) 

plt.show()
```

<i>style</i> <code>ggplot</code>.

```python
# Mengatur style 

plt.style.use('ggplot')

X = np.linspace(0, 30, 100)
y1 = np.cos(X*2)
y2 = np.cos(X)
y3 = np.sin(X/3)
y4 = np.sin(X)

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(X, y1)
ax.plot(X, y2)
ax.plot(X, y3) 
ax.plot(X, y4) 

plt.show()
```

## Menyimpan gambar

Untuk menyimpan gambar grafik yang telah dibuat, dapat menggunakan <i>method</i> <code>savefig()</code> yang diberikan argumen berupa nama <i>file</i> gambar yang ingin disimpan. Misalnya dari contoh sebelumnya, ingin disimpan dengan nama <code>contoh_visualisasi.png</code>. Perhatikan kode berikut.

```python
plt.style.use('ggplot')

X = np.linspace(0, 30, 100)
y1 = np.cos(X*2)
y2 = np.cos(X)
y3 = np.sin(X/3)
y4 = np.sin(X)

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(X, y1)
ax.plot(X, y2)
ax.plot(X, y3) 
ax.plot(X, y4) 

# Menyimpan gambar
fig.savefig('contoh_visualisasi.png')
```

Setelah cek di folder yang sama dengan yang digunakan untuk menyimpan <i>file</i> notebook ini, <i>file</i> <code>contoh_visualisasi.png</code> otomatis tersimpan disana.
