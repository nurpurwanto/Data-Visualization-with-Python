<i>load</i> dataset dengan <code>read_csv()</code> dan menyimpannya di dalam variabel <code>df</code>. Dataset ini belum memiliki nama kolom yang jelas, sehingga perlu dibuat nama kolom yang mewakili data dari masing-masing kolom.

```python
# list nama kolom
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
              'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
              'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower',
              'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

# read dataset
df = pd.read_csv('automobile.data', names=column_names)

df.head()
```

Ada beberapa <i>cell</i> yang mengandung <i>missing values</i> berupa simbol tanda tanya <code>?</code>. Oleh karena itu, ubah dulu tanda <code>?</code> menjadi __NaN__.

```python
# replace missing values '?' dengan NaN

df = df.replace('?', np.nan)
df.head()
```

Kemudian hapus data yang mengandung NaN pada kolom <code>price</code> menggunakan <code>dropna()</code>. Sementara itu, untuk kolom lainnya, ganti nilai NaN dengan nilai tertentu seperti di bawah ini.

```python
# Menghapus baris yang mengandung mising values di kolom price
df.dropna(subset=['price'], axis=0, inplace=True)

# Me-reset indeks karena ada data yang terhapus
df.reset_index(drop=True, inplace=True)

# Mengganti missing values dengan 'four' untuk kolom num-of-doors
df['num-of-doors'].fillna('four', inplace=True)

# Mengganti missing values dengan mean-nya untuk kolom lainnya
avg_norm = df['normalized-losses'].astype('float').mean(axis=0)
df['normalized-losses'].replace(np.nan, avg_norm, inplace=True)

avg_stroke = df['stroke'].astype('float').mean(axis=0)
df['stroke'].replace(np.nan, avg_stroke, inplace=True)

avg_bore = df['bore'].astype('float').mean(axis=0)
df['bore'].replace(np.nan, avg_bore, inplace=True)

avg_horse = df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_horse, inplace=True)

avg_peak = df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_norm, inplace=True)

df.head()
```

Selanjutnya kita ganti tipe datanya.

```python
# Konversi tipe data

df[['bore', 'stroke', 'peak-rpm']] = df[['bore', 'stroke', 'peak-rpm']].astype('float')
df[['normalized-losses', 'horsepower']] = df[['normalized-losses', 'horsepower']].astype('int')
df[['price']] = df[['price']].astype('float')
```

cek info dataframe <code>df</code> setelah penanganan <i>missing values</i>.

```python
# Mengecek info dataframe

df.info()
```

__Heatmap__ adalah salah satu jenis visualisasi data yang memetakan kolom-kolom pada dataframe ke dalam axis X dan Y sebagai kotak-kotak berwarna. Heatmap lebih mudah dibuat menggunakan <i>library</i> __Seaborn__ karena memiliki <i>method</i> tersendiri, yaitu <code>heatmap()</code>, untuk men-<i>generate</i>-nya.
    
__Seaborn__ merupakan salah satu <i>library</i> visualisasi data sama seperti Matplotlib. Perbedaan pada fungsionalitasnya adalah Matplotlib baik digunakan untuk membuat pola-pola dasar, sedangkan Seaborn baik digunakan jika ingin membuat variasi pola visualisasi.

Siapkan dahulu dataframenya. Untuk lebih mudah, lakukan <i>reshaping</i> terlebih dahulu terhadap dataframe <code>df</code> menggunakan <code>pivot_table()</code>. 

Misal jadikan <code>drive-wheels</code> sebagai indeksnya, <code>body-style</code> sebagai kolomnya, dan <code>price</code> sebagai nilainya. Untuk <i>aggregate function</i>-nya gunakan <i>mean</i>. Jadi nanti yang tampil sebagai nilainya adalah <i>mean</i> dari harganya.

```python
# Membuat pivot tabel

df_pivot = df.pivot_table(index='drive-wheels', columns='body-style', values='price', aggfunc='mean')
df_pivot
```

Setelah itu, dengan menggunakan <i>method</i> <code>heatmap()</code> dari Seaborn. Perhatikan kode berikut. 

```python
# Membuat heatmap

sns.heatmap(df_pivot)
```

Dua kotak di sebelah kiri atas tidak terbentuk (berwarna putih) karena memang nilai tersebut adalah __NaN__.

wokay kemudian percantik heatmap tersebut dengan mengatur ukuran <i>figure</i> maupun label. Untuk itu, kita dapat menggunakan Matplotlib.

```python
# Menggunakan Matplotlib untuk mengatur plot

plt.figure(figsize=(12,7))    # Mengatur ukuran figure

# Membuat heatmap
sns.heatmap(df_pivot)

plt.title('Harga Rata-rata Mobil Berdasarkan Body Style dan Drive Wheels\n', size=16)
plt.xlabel('\nBody Style', size=14)
plt.ylabel('Drive Wheels\n', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.show()
```

### Menambahkan teks label

Menambahkan teks pada kotak-kotak tersebut. Teks disini maksudnya nilai yang ada pada dataframe, dalam hal ini nilai rata-ratanya. 

Untuk menampilkan teks, kita cukup menambahkan argumen <code>annot = True</code> pada <code>heatmap()</code>. Karena nilai yang ada di dataframe <code>df_pivot</code> tersebut panjang (angka di belakang koma ditulis semua), atur jumlah angka di belakang koma. Disini kita tambahkan argumen <code>fmt='.2f'</code> yang berarti hanya menampilkan 2 angka di belakang koma desimal.

```python
# Menambahkan teks pada heatmap

plt.figure(figsize=(12,7))    # Mengatur ukuran figure

# Membuat heatmap dengan teks
sns.heatmap(df_pivot, annot=True, fmt='.2f')

plt.title('Harga Rata-rata Mobil Berdasarkan Body Style dan Drive Wheels\n', size=16)
plt.xlabel('\nBody Style', size=14)
plt.ylabel('Drive Wheels\n', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.show()
```

### Menambahkan label

Untuk menambahkan label pada heatmap menggunakan Matplotlib agak rumit karena harus menambahkan beberapa baris kode. Pada contoh ini, akan memisahkan blok kode untuk menampilkan label atau nilai pada dataframe pada heatmap dengan membuat fungsi tersendiri yang dapat dipanggil pada blok kode utama. Buat fungsi bernama <code>show_values()</code> seperti di bawah ini.

```python
def show_values(pc, fmt="%.2f", **kw):
    pc.update_scalarmappable()
    ax = pc.axes
    for p, color, value in zip(pc.get_paths(), pc.get_facecolors(), pc.get_array()):
        x, y = p.vertices[:-2, :].mean(0)
        if np.all(color[:3] > 0.5):
            color = (0.0, 0.0, 0.0)
        else:
            color = (1.0, 1.0, 1.0)
        ax.text(x, y, fmt % value, ha="center", va="center", color=color, **kw)
```

Buat blok kode utama untuk menampilkan heatmap dengan Matplotlib dan memanggil fungsi <code>show_values()</code>. Perhatikan kode berikut.

```python
# Menambahkan label/values

fig, ax = plt.subplots(figsize=(12,7))

# Membuat heatmap dengan pcolor()
heatmap = ax.pcolor(df_pivot, cmap='Reds')

fig.colorbar(heatmap)  # Menampilkan colorbar
show_values(heatmap)   # Menampilkan teks/label/values
    
plt.title('Harga Rata-rata Mobil Berdasarkan Body Style dan Drive Wheels\n', size=16)

plt.xlabel('\nBody Style', size=14)
plt.ylabel('Drive Wheels\n', size=14)

plt.xticks(np.arange(0.5, len(df_pivot.columns), 1), df_pivot.columns, size=14)     # Menampilkan xticklabel
plt.yticks(np.arange(0.5, len(df_pivot.index), 1), df_pivot.index, size=14)       # Menampilkan yticklabel
plt.show()
```

Sekarang heatmap-nya sudah berlabel.
