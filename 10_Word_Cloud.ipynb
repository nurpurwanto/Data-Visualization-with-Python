## Apa itu Wordcloud ?

__Word Cloud__ adalah salah satu visualisasi data yang digunakan untuk melihat kata-kata yang paling sering muncul di sebuah data teks seperti artikel, buku, dan lain sebagainya. Semakin sering kata tersebut muncul, maka akan semakin besar tulisan kata tersebut pada <i>Word Cloud</i>.

Di Python, membuat <i>Word Cloud</i> menggunakan <i>package</i> <code>wordcloud</code> yang sudah tersedia. 

### Install Wordcloud

Untuk membuat __Word Cloud__ harus meng-<i>install</i> <i>package</i>-nya terlebih dahulu. Jalankan kode di bawah untuk menginstal <code>wordcloud</code>.

```python
pip install wordcloud
```

Setelah terinstal, <i>import</i>-nya sebelum menggunakannya. Disini kita mengimpor beberapa <i>library</i> atau <i>package</i> seperti Matplotlib, Numpy, Pandas, dan tentunya WordCloud. juga harus mengimpor __STOPWORDS__ karena akan kita gunakan untuk mem-<i>filter</i> kata-kata yang kurang penting pada string.

```python
# import package yang dibutuhkan

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Membuka file dan memasukkannya ke dalam variabel novel_alice

novel_alice = open('novel-alice.txt', 'r', encoding='utf8').read()
```

Melihat isi dari teks tersebut dengan <code>print</code>.

```python
# Menampilkan teks novel_alice

print(novel_alice)
```

Mendefinisikan <i>stopwords</i> dari <i>package</i> <code>wordcloud</code> yang telah <i>diimport</i> sebelumnya. <i>Stopwords</i> yang tersedia di Python adalah versi bahasa inggris. Untuk <i>stopwords</i> berbahasa selain bahasa inggris.

```python
# Mendefinisikan stopwords

stopwords = set(STOPWORDS)
```

Sekarang <code>print</code> stopwords.

```python
# Menampilkan stopwords

print(stopwords)
```

## Visualisasi teks dengan Word Cloud

Untuk membuat <i>Word Cloud</i>, pertama harus membuat objek <code>WordCloud</code> dan men-<i>generate</i>-nya dengan <i>method</i> <code>generate()</code>. Perhatikan kode berikut.

```python
# Membuat objek word cloud
wordcloud_alice = WordCloud(
    background_color='white',
    max_words=10000,
    stopwords=stopwords
)

# Generate word cloud
wordcloud_alice.generate(novel_alice)
```

Pada objek <i>word cloud</i>, dapat diberikan beberapa argumen diantaranya yang digunakan pada contoh kode di atas adalah :

* <code>background_color</code> untuk mengatur warna latar belakang. Pada contoh di atas, definisikan warna putih (white) sebagai warna latarnya.
* <code>max_words</code> untuk mendefinisikan jumlah kata yang akan diambil. Pada contoh di atas, hanya 10000 kata pertama dari novel_alice yang akan digunakan untuk membuat word cloud.
* <code>stopwords</code> untuk mem-filter kata-kata tertentu agar tidak muncul pada hasil visualisasi.

Selanjutnya, tampilkan objek <i>word cloud</i> tersebut menggunakan <i>method</i> dari Matplotlib yaitu <code>imshow()</code>.

```python
# Menampilkan word cloud

plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Mengatur ukuran figure

<i>word cloud</i> yang terbentuk berukuran kecil. Untuk menentukan ukuran <i>figure</i>, atur dengan <i>method</i> <code>set_figwidth()</code> dan <code>set_figheight()</code>.

```python
fig = plt.figure()
fig.set_figwidth(14) # mengatur width
fig.set_figheight(10) # mengatur height

# menampilkan word cloud
plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

*Atau*, gunakan <code>figsize</code> seperti di bawah ini.

```python
# Mengatur ukuran figure

plt.figure(figsize=(14,10))
plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Menghapus kata pada Word Cloud

Menghilangkan salah satu kata pada <i>word cloud</i> karena mungkin dianggap kurang penting. Untuk itu, harus memasukkan kata tersebut ke dalam <i>stopwords</i> agar tidak diikutsertakan dalam pembuatan <i>word cloud</i>.

Misalnya lihat dari output <i>word cloud</i> sebelumnya, ada string __'S'__ yang tidak bermakna apapun pada <i>word cloud</i> dan ingin dihapus. Tambahkan __'S'__ pada <i>stopwords</i>. Perhatikan kode berikut.

```python
stopwords.add('S') # menambahkan 'S' pada stopwords

# men-generate ulang word cloud
wordcloud_alice.generate(novel_alice)

# menampilkan word cloud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(10) 

plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Sekarang <i>word cloud</i> tersebut sudah tidak mengandung string __'S'__.

Bagaimana jika ingin menambahkan beberapa kata pada <i>stopwords</i> ?

Tambahkan __for__ <i>looping</i> seperti di bawah ini.

```python
add_stopword = ['don', 't', 'without', 'll', 'tm']   # membuat list stopwords

# menambahkan add_stopword pada stopwords
for i in add_stopword:
    stopwords.add(i) 

# men-generate ulang word cloud
wordcloud_alice.generate(novel_alice)

# menampilkan word cloud
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(10) 

plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Visualisasi data pada dataframe dengan Word Cloud

Normalnya, <i>word cloud</i> di-<i>generate</i> dari data teks atau string. Namun, bagaimana jika ingin membuat <i>word cloud</i> dari data pada dataframe? Untuk melakukan hal itu, salah satu caranya yaitu membuat data tersebut menjadi string.

Misal menggunakan dataset populasi penduduk di setiap negara di dunia, dan  ingin menampilkan nama-nama negara dalam bentuk <i>word cloud</i> berdasarkan pada jumlah populasinya. Semakin banyak populasinya, maka nama negara tersebut akan semakin besar di <i>word cloud</i>.

Pertama,  <i>load</i> datasetnya terlebih dahulu dengan <code>read_csv()</code> dan kemudian kita simpan pada variabel <code>df</code>.

```python
# Load dataset

df = pd.read_csv("pop_worldometer_data.csv")
df.head(10)
```

Kemudian agar lebih mudah dalam <i>indexing</i>, atur kolom <code>Country (or dependency)</code> sebagai indeks menggunakan <i>method</i> <code>set_index()</code>.

```python
# Mengatur 'Country (or dependency)' sebagai indeks

df.set_index("Country (or dependency)", inplace=True)
```

lihat dataframenya setelah perubahan dilakukan.

```python
# Menampilkan 10 data teratas

df.head(10)
```

### Mengubah data pada dataframe menjadi string

Untuk membuat <i>word cloud</i> dari dataframe, cara pertama yang dapat dilakukan adalah dengan mengubah data pada dataframe menjadi string. Dalam kasus ini, jadikan nama-nama negara (indeks dataframe) sebagai string dengan masing-masing negara memiliki jumlah tertentu. Perhatikan kode berikut.

```python
# Membuat string dari indeks dataframe 

word_string = ''      # Mendefinisikan string kosong

# Perhitungan untuk menentukan jumlah untuk tiap negara
for country in df.index.values:
    repeat_num_times = int(df.loc[country, 'World Share %']*100)
    word_string = word_string + ((country + ' ') * repeat_num_times)
                                     
# display the generated text
word_string
```

Dari kode di atas, tentukan jumlah kata untuk tiap negara yang tercetak dengan mengalikan kolom <code>World Share %</code> dengan __100__. 

Setelah memiliki string-nya, buat objek <i>word cloud</i>, men-<i>generate</i>-nya, dan menampilkannya dengan Matplotlib seperti di bawah ini.

```python
# Membuat word cloud
wordcloud_country = WordCloud(background_color='black', 
                              collocations=False, 
                              colormap='viridis',
                              width=600,
                              height=330)

wordcloud_country.generate(word_string)

plt.figure(figsize=(16,12))
plt.imshow(wordcloud_country, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Ada beberapa parameter yang baru, di antaranya:

* <code>collocations</code> : atur dengan nilai __False__ agar kemunculan nama negara tidak <i>double</i>.
* <code>colormap</code> : untuk mengatur warna teks pada <i>word cloud</i>.
* <code>width</code> : untuk mengatur lebar (horizontal) <i>word cloud</i>.
* <code>height</code> : untuk mengatur tinggi (vertikal) <i>word cloud</i>.

### Membuat Word Cloud dari frekuensi kata

Cara yang kedua untuk membuat <i>word cloud</i> dari dataframe adalah dari frekuensi setiap kata. Disini tidak membuat string atau teks dari dataframe, melainkan membuat 2 list dengan jumlah elemen yang sama, dimana satu list berfungsi untuk menampung kata (nama negara) dan satu list lagi untuk menampung jumlah frekuensi. Untuk menghasilkan frekuensinya masih menggunakan perhitungan hasil kali antara kolom <code>World Share %</code> dengan __100__.

```python
# Membuat list_country dan list_freq

list_country = []     # Mendefinisikan list kosong untuk menampung nama negara
list_freq = []        # Mendefinisikan list kosong untuk menampung frekuensi

for country in df.index.values:
    frequency = int(df.loc[country, 'World Share %']*100)
    
    list_country.append(country)         # Menambahkan nama negara ke list_country
    list_freq.append(frequency)          # Menambahkan frekuensi ke list_freq
```

Membuat dictionary dari 2 list tersebut. Misal membuat dictionary dan memasukkannya ke dalam variabel <code>d</code>.

```python
# Membuat dictionary

d = dict(zip(list_country, list_freq))
print(d)
```

Membuat objek <i>word cloud</i> dan men-<i>generate</i>-nya. 

Ada yang berbeda dalam proses <i>generate</i> ini. yakni menggunakan <i>method</i> <code>generate_from_frequencies()</code> untuk men-<i>generate</i> <i>word cloud</i> dari frekuensinya.

```python
# Membuat objek word cloud
wordcloud_country = WordCloud(background_color='purple', 
                              collocations=False, 
                              colormap='Paired',
                              width=600,
                              height=330)

# Generate word cloud from frequencies
wordcloud_country.generate_from_frequencies(d)

plt.figure(figsize=(16,12))
plt.imshow(wordcloud_country, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Menambahkan kolom 'frequency' pada dataframe

Selain dengan membuat 2 list seperti di atas, juga dapat melakukannya dengan cara lain yaitu dengan menambahkan kolom baru yang berisi frekuensinya. 
Buat kolom baru yaitu <code>Frequency for Word Cloud</code> yang berisi hasil kali dari kolom <code>World Share %</code> dan __100__.

```python
# Membuat kolom baru 'Frequency for Word Cloud'

df['Frequency for Word Cloud'] = df['World Share %'] * 100
df.head()
```

Buat dictionary dari <code>df.index.values</code> dan <code>df['Frequency for Word Cloud']</code> dan men-<i>generate</i> <i>word cloud</i> dengan <code>generate_from_frequencies()</code> dengan memasukkan argumen berupa variabel <code>d</code> atau dictionary tersebut.

```python
d = dict(zip(df.index.values, df['Frequency for Word Cloud']))

# Create the word cloud
wordcloud_country = WordCloud(background_color='maroon', 
                              collocations=False, 
                              colormap='coolwarm',
                              width=600,
                              height=330)

# Generate word cloud from frequencies
wordcloud_country.generate_from_frequencies(d)

plt.figure(figsize=(16,12))
plt.imshow(wordcloud_country, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Word Cloud pada berbagai bentuk objek

Tidak hanya <i>word cloud</i> berbentuk persegi, Bisa juga dibuat <i>word cloud</i> dengan berbagai bentuk. Jadi disini akan dibuat <i>word cloud</i> pada gambar tertentu yang menjadi <i>mask</i>-nya.

Karena akan menggunakan gambar, kita harus meng-<i>import</i> <code>Image</code> dari <i>package</i> <code>PIL</code>. Gambar yang adigunakan adalah <code>alice_mask.png</code>.

```python
from PIL import Image

# Menyimpan mask pada variabel alice_mask
alice_mask = np.array(Image.open('images/alice_mask.png'))
```

Selanjutnya tampilkan gambar atau <i>mask</i> tersebut dengan <code>imshow()</code>.

```python
# Menampilkan gambar mask

fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Kemudian buat <i>word cloud</i>-nya sama seperti sebelumnya namun ditambah parameter <code>mask</code> yang diberi nilai nama variabel dari <i>mask</i>-nya.

```python
# Membuat objek word cloud
wordcloud_alice = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)

# Generate word cloud
wordcloud_alice.generate(novel_alice)

# Menampilkan word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Sekarang gunakan <code>colormap</code> dengan tipe __coolwarm__.

```python
# Membuat objek word cloud
wordcloud_alice = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords, colormap='coolwarm')

# Generate word cloud
wordcloud_alice.generate(novel_alice)

# Menampilkan word cloud
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 

plt.imshow(wordcloud_alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```
